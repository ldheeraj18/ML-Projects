```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Ames_Housing_Data_Final.csv')
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0', axis = 1)
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.drop('SalePrice', axis = 1)
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.1, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_X_train = scaler.transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elastic_net_model = ElasticNet()
```

```python
e_param = {
    'alpha': [0.1,1,5,10,50,100],
    'l1_ratio': [.1, .5 , .7 , .9 , .95 , .99 , 1]
}
```

```python
grid_elastic_net = GridSearchCV(estimator = elastic_net_model,param_grid=e_param, scoring = 'neg_mean_squared_error', cv = 5, verbose=1)
```

```python
import warnings
warnings.filterwarnings('ignore')
grid_elastic_net.fit(scaled_X_train, y_train)
```

```python
grid_elastic_net.best_params_
```

```python
predictions = grid_elastic_net.predict(scaled_X_test)
```

```python
predictions
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test , predictions)
MAE
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, predictions))
RMSE
```
